```python
import requests
import pandas as pd
import csv
from tqdm.auto import tqdm
from functools import reduce
```

```python
objects_data = {}
offset_ = 0
limit = 100
objIds = []

while offset_ <= 10500:
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit=100&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = requests.get(url)
    res.json()
    objects_data = res.json()
    objects_list = objects_data.get('data').get('list')
    objects_list
    
    objIds_temp = [x.get('objId') for x in objects_list]
    
    objIds.append(objIds_temp)
    
    offset_ += 100
```

```python
len(objIds)
```

```python
objIds
```

```python
objIds_single = reduce(lambda x,y: x+y, objIds)
```

```python
d = []
for obj in tqdm(objIds_single):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{obj}'
    res = requests.get(url)
    d.append(res.json())
pd.json_normalize(d)
```

```python
df = pd.json_normalize(d)
```

```python
df.to_excel("nashdom.xlsx")
```
