```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, RidgeCV, LassoCV, LogisticRegression
from sklearn.dummy import DummyClassifier
```

```python
RANDOM_STATE = 42
```

```python
dataset = load_boston()
X = pd.DataFrame(dataset.data)
X.columns = dataset.feature_names
y = dataset.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = RANDOM_STATE)
```

3. Для Ridge и Lasso подберите коэффициент регуляризации(используйте GridSearchCV, RidgeCV, LassoCV) в пределах от $10^{-5}$ до $10^5$ (по степеням 10). Посчитайте R2 на тестовой выборке по лучшим моделям и сравните с предыдущими результатами. Напишите как изменился результат

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
```

```python
#GridSearchCV для Ridge

params = {'alpha': [10 ** n for n in range(-5, 6)]}
grd_cv_ridge = GridSearchCV (estimator = Ridge(), param_grid = params)
grd_cv_ridge.fit (X_train, y_train)
y_pred = grd_cv_ridge.predict(X_test)
r2 = r2_score (y_test, y_pred)
print (f"R2 for GridSearchCV ridge: {r2:.6f}")

print (grd_cv_ridge.best_estimator_)




##GridSearchCV для Lasso

params = {'alpha': [10 ** n for n in range(-5, 6)]}
grd_cv_lasso = GridSearchCV (estimator = Lasso(), param_grid = params)
grd_cv_lasso.fit (X_train, y_train)
y_pred = grd_cv_lasso.predict(X_test)
r2 = r2_score (y_test, y_pred)
print (f"R2 for GridSearchCV lasso: {r2:.6f}")

print (grd_cv_lasso.best_estimator_)




#RidgeCV для Ridge

params = {'alpha': [10 ** n for n in range(-5, 6)]}
ridge_cv = RidgeCV (alphas = params['alpha'])
ridge_cv.fit (X_train, y_train)
y_pred = ridge_cv.predict(X_test)
r2 = r2_score (y_test, y_pred)
print (f"R2 for ridge_cv: {r2:.6f}")

print (f"RidgeCV for Ridge (alpha = {ridge_cv.alpha_:.6f})")




#LassoCV для Lasso

params = {'alpha': [10 ** n for n in range(-5, 6)]}
lasso_cv = LassoCV (alphas = params['alpha'])
lasso_cv.fit (X_train, y_train)
y_pred = lasso_cv.predict(X_test)
r2 = r2_score (y_test, y_pred)
print (f"R2 for lasso_cv: {r2:.6f}")

print (f"LassoCV for Lasso (alpha = {lasso_cv.alpha_:.6f})")
```

```python
lin_svc_pipe = Pipeline (steps = [('transformer', transformer), ('LinearSVC', LinearSVC(random_state = RANDOM_STATE))])
scores_ac = cross_val_score (lin_svc_pipe, X, y, scoring='accuracy')
scores_f1 = cross_val_score (lin_svc_pipe, X, y, scoring='f1')
print('lin_svc accuracy: ', scores_ac.mean())
print('lin_svc f1: ', scores_f1.mean())
```
